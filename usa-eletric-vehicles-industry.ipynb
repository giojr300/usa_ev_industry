# 1. Dataset Analysing

This section will analyse the basics characteristics of the dataset, such as:

```python
vehicle.info()
```

```python
print(f'Lines: {vehicle.shape[0]} | Columns: {vehicle.shape[1]}')
```

```python
vehicle.isnull().sum()
```

```python
vehicle = vehicle.drop(columns=['Vehicle Location', 
                                'Electric Utility', 
                                'Legislative District', 
                                'Postal Code', 
                                '2020 Census Tract',
                                'DOL Vehicle ID',
                                'Base MSRP'])
```

```python
for c in vehicle.select_dtypes(include='object'):
    print(f'{c}: {vehicle[c].unique()}')
    print('******')
```

```python
vehicle = vehicle.rename(columns={'VIN (1-10)': 'Id',
                        'Make': 'Company', 
                        'Clean Alternative Fuel Vehicle (CAFV) Eligibility': 'Clean Alternative Fuel Vehicle Eligibility'})
```

# 2. Data Visualization
 
This section will be dedicated to plot ghaps and understand the following questions:

- What are the leaders companys from EVs in USA;
- How American EVs industry has been growing through the years;
- Comparison from EVs types distribution.
- Top 10 best-selling EVs models.

```python
company = vehicle.groupby('Company').agg({'Id': 'count'}).sort_values('Id', ascending=False).reset_index()
company = company.rename(columns={'Id': 'Count'})

fig = go.Figure()
fig.add_trace(go.Bar(x=company.Company, y=company.Count))
fig.update_xaxes(title_text='Company')
fig.update_yaxes(title_text='EVs Unit Sales')
fig.update_layout(title='Leaders Companys from EVs in USA')
fig.show() 
```

```python
model_year = vehicle.groupby('Model Year').agg({'Id': 'count'}).reset_index()
model_year = model_year.rename(columns={'Id': 'Count'})

fig = go.Figure()
fig.add_trace(go.Line(x=model_year['Model Year'], y=model_year['Count']));
fig.update_xaxes(title_text="Year")
fig.update_yaxes(title_text='EVs Unit Sales')
fig.update_layout(title='EVs Sales Through the Year')
fig.show();
```

```python
fuel = vehicle.groupby('Electric Vehicle Type')['Id'].count().reset_index()
fuel = fuel.rename(columns={'Electric Vehicle Type': 'Type', 'Id': 'Count'})

fig = go.Figure()
fig.add_trace(go.Pie(values=fuel['Count'], labels=fuel['Type']))
fig.update_layout(title='Eletric Vehicle Types Distribution')
fig.show()
```

```python
top10_model = vehicle.groupby(['Company', 'Model']).agg({'Id': 'count'}).reset_index().sort_values('Id', ascending=False)[:10]
top10_model = top10_model.rename(columns={'Id': 'Count'})

fig = go.Figure()
fig.add_trace(go.Bar(y=top10_model.Model, x=top10_model.Count, orientation='h',))
fig.update_yaxes(title_text='Model')
fig.update_xaxes(title_text='Sales (Unit)')
fig.update_layout(title='Top 10 Best-Selling EVs Model')
fig.show()
```
